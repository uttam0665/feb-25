```python
'''Q2. What is the difference between the functions loc and iloc in pandas.DataFrame?
Ans:-

In pandas, the functions loc and iloc are used to select data from a DataFrame based on label or integer-location, respectively.
loc is used to select data based on labels or boolean conditions, while iloc is used to select data based on integer locations.
Here is an example to illustrate the difference between the two:'''   
    
import pandas as pd

df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6], 'C': [7, 8, 9]}, index=['a', 'b', 'c'])
print(df.loc['a'])  
print(df.iloc[0]) 
print(df.iloc[0:2])  
print(df.loc['a':'b'])  


In summary, loc is used for label-based indexing and slicing, while iloc is used for integer-based indexing and slicing.



```

```python
Q3. Reindex the given dataframe using a variable, reindex = [3,0,1,2] and store it in the variable, new_df
then find the output for both new_df.loc[2] and new_df.iloc[2].

Did you observe any difference in both the outputs? If so then explain it.
Consider the below code to answer further questions:
import pandas as pd
import numpy as np
columns = ['column_1', 'column_2', 'column_3', 'column_4', 'column_5', 'column_6']
indices = [1,2,3,4,5,6]
#Creating a dataframe:
df1 = pd.DataFrame(np.random.rand(6,6), columns = columns, index = indices)

Ans:-
    
To reindex the given dataframe using a variable reindex = [3,0,1,2], we can use the reindex method of the dataframe object and pass the reindex list to it. The resulting dataframe will have the rows arranged in the order specified in the reindex list. We can store this new dataframe in the variable new_df.
Here's the code to achieve this  

import pandas as pd
import numpy as np
columns = ['column_1', 'column_2', 'column_3', 'column_4', 'column_5', 'column_6']
indices = [1,2,3,4,5,6]
df1 = pd.DataFrame(np.random.rand(6,6), columns = columns, index = indices)
reindex = [3,0,1,2]
new_df = df1.reindex(reindex)
print(new_df.loc[2])
print(new_df.iloc[2])


The output of new_df.loc[2] and new_df.iloc[2] will be different because they retrieve the data based on different indexing methods.
new_df.loc[2] retrieves the row with label 2 in the new index, which corresponds to the row with label 3 in the original index (since the original row with label 2 was moved to position 0 in the new index).
Therefore, the output of new_df.loc[2] will be the row with label 2 in the original dataframe, but with the order of columns shuffled according to the new index.
new_df.iloc[2] retrieves the third row in the new index (which corresponds to the row with label 1 in the original index, since it was moved to position 2 in the new index). 
Therefore, the output of new_df.iloc[2] will be the third row of the original dataframe, but with the order of columns shuffled according to the new index.

Further, if we want to retrieve the data for the third row in the original dataframe, regardless of the new index order, we can use df1.iloc[2], which will return the third row of the original dataframe.



```

```python
Q6. What do you understand about the windows function in pandas and list the types of windows
functions?
Ans:-

In Pandas, a window function (also known as a rolling or sliding window function) is a way to perform calculations over a specified window of data, where the window moves over the data frame. 
It is useful for calculating running or cumulative statistics and smoothing out noisy data.
Pandas provides several types of window functions, including:

1.Rolling - It calculates the statistic over a sliding window of a fixed size.

2.Expanding - It calculates the statistic for all the data points up to the current point.

3.Exponentially Weighted Moving Average (EWMA) - It calculates the exponentially weighted moving average of the data.

4.Rolling.apply - It applies a custom function to a rolling window.

5.Rolling.aggregate - It applies multiple aggregation functions to a rolling window.

5.Rolling.agg - It applies a specific aggregation function to a rolling window.

6.Rolling.count - It counts the number of non-null values in a rolling window.

7.Rolling.cov - It calculates the covariance of two series over a rolling window.

8.Rolling.max - It calculates the maximum value in a rolling window.

9.Rolling.mean - It calculates the mean value in a rolling window.

10.Rolling.median - It calculates the median value in a rolling window.

11.Rolling.min - It calculates the minimum value in a rolling window.

12.Rolling.quantile - It calculates the specified quantile value in a rolling window.

13.Rolling.skew - It calculates the skewness of the data in a rolling window.

14.Rolling.std - It calculates the standard deviation of the data in a rolling window.

15.Rolling.sum - It calculates the sum of the data in a rolling window.

16.Rolling.var - It calculates the variance of the data in a rolling window.

These window functions can be applied to a Pandas DataFrame or Series using the .rolling() method.



    
```

```python
'''Q8. Write a Python program that takes in two dates as input (in the format YYYY-MM-DD) and
calculates the difference between them in days, hours, and minutes using Pandas time delta. The
program should prompt the user to enter the dates and display the result.
Ans:-

Here's the Python program that takes two dates as input from the user and calculates the difference between them in days, hours, and minutes using Pandas timedelta:'''

import pandas as pd
date1 = input("Enter the first date (YYYY-MM-DD): ")
date2 = input("Enter the second date (YYYY-MM-DD): ")
date1 = pd.to_datetime(date1)
date2 = pd.to_datetime(date2)
delta = date2 - date1
days = delta.days
hours = delta.seconds // 3600
minutes = (delta.seconds % 3600) // 60
print(f"The difference between {date1} and {date2} is {days} days, {hours} hours, and {minutes} minutes.")

In this program, we first import the Pandas library and prompt the user to enter two dates in the format YYYY-MM-DD.
We then convert the input strings to datetime objects using the pd.to_datetime() function. Next, we calculate the timedelta between the two dates using the - operator.
To extract the number of days, hours, and minutes from the timedelta, we use the days, seconds, and minutes attributes of the timedelta object. 
We convert the seconds to hours and minutes using integer division (//) and modulo (%) operators.

Finally, we use an f-string to display the result to the user. The f-string includes the original dates entered by the user, as well as the calculated difference in days, hours, and minutes.
```

```python
"""Q9. Write a Python program that reads a CSV file containing categorical data and converts a specified
column to a categorical data type. The program should prompt the user to enter the file path, column
name, and category order, and then display the sorted data."""
Ans:-

Sure, here's an example Python program that reads a CSV file containing categorical data and converts a specified column to a categorical data type.
It then prompts the user to enter the file path, column name, and category order, and displays the sorted data.    

import pandas as pd
file_path = input("Enter file path: ")
column_name = input("Enter column name: ")
category_order = input("Enter category order (comma-separated): ").split(",")
df = pd.read_csv(file_path)
df[column_name] = pd.Categorical(df[column_name], categories=category_order, ordered=True)
df_sorted = df.sort_values(column_name)
print(df_sorted)
```

```python
'''Q10. Write a Python program that reads a CSV file containing sales data for different products and
visualizes the data using a stacked bar chart to show the sales of each product category over time. The
program should prompt the user to enter the file path and display the chart.
Ans:-

Sure, here's a Python program that reads a CSV file containing sales data for different products and visualizes the data using a stacked bar chart to show the sales of each product category over time:'''

import pandas as pd
import matplotlib.pyplot as plt
file_path = input("Enter file path: ")
df = pd.read_csv(file_path)
grouped_data = df.groupby(['Product Category', 'Year'])['Sales'].sum().unstack()
ax = grouped_data.plot(kind='bar', stacked=True, figsize=(10, 6))
ax.set_title('Sales by Product Category')
ax.set_xlabel('Year')
ax.set_ylabel('Sales')
plt.show()


This program uses the Pandas library to read the CSV file and group the data by product category and year, and the Matplotlib library to create the stacked bar chart.
The user is prompted to enter the file path, and the program displays the chart after reading and processing the data.
```

```python
'''Q11. You are given a CSV file containing student data that includes the student ID and their test score. Write
a Python program that reads the CSV file, calculates the mean, median, and mode of the test scores, and
displays the results in a table.
Ans:-


Here's the Python program that reads a CSV file containing student data, calculates the mean, median, and mode of the test scores, and displays the results in a table:'''

import pandas as pd
csv_path = input("Enter the file path of the CSV file containing the student data: ")
df = pd.read_csv(csv_path)
mean_score = df["Test Score"].mean()
median_score = df["Test Score"].median()
mode_score = df["Test Score"].mode()
table = pd.DataFrame({
    "Statistic": ["Mean", "Median", "Mode"],
    "Value": [mean_score, median_score, mode_score]
})

print(table)


If the CSV file student_data.csv is located in the same directory as the Python program, you can enter student_data.csv as the file path when prompted. 
The output should be:

Note that the mode is returned as a Series object, which contains all the modes of the data. 
In this example, the test scores 85 and 90 both occur twice, so they are both modes of the data.

    
```
